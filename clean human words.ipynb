```python
import pandas as pd
import re
```

```python
filename = 'result_2020_11_15.json'
```

```python
df = pd.read_json(filename)
```

```python
pattern = re.compile(r"[^a-zA-Z#\’']")

def process_labels(row):
    
    data = row['label_text']
    
    #remove brackets
    data = data[1:-1]
    
    # replace \n\n with ,
    data = data.replace('\n\n',',')
    
    # run regexp
    data = pattern.sub(',',data)
    
    #split on ,
    data = data.split(',')

    #lower case and remove empty words
    return [x.lower() for x in data if x != '']    
```

```python
df['human_words'] = df.apply(process_labels,axis=1)
```
